# **Варианты исполнения команд**

---

### **1. Какие существуют способы запуска Python-кода?**
**Ответ:**
Основные способы запуска:
1. **Интерактивный режим** (REPL) — ввод команд напрямую в интерпретаторе через `python` в терминале.
2. **Запуск скрипта** — выполнение файла `.py` командой `python script.py`.
3. **Командная строка** — однострочные команды через `python -c "print('Hello')"`.

---

### **2. В чём разница между интерактивным и файловым режимом?**
**Ответ:**
- **Интерактивный режим** (REPL):
  - Команды выполняются сразу после ввода.
  - Подходит для тестирования кода, экспериментов.
  - Нет сохранения истории (если не использовать `IPython` или `Jupyter`).

- **Файловый режим** (скрипт):
  - Код сохраняется в файл `.py` и выполняется целиком.
  - Можно редактировать и перезапускать.
  - Подходит для сложных программ и проектов.

**Пример:**
```bash
# Интерактивно:
>>> print("Hello")  # Выводится сразу

# В файле script.py:
print("Hello")  # Запускается через python script.py
```

---

### **3. Как запустить Python-скрипт из командной строки/терминала?**
**Ответ:**
1. Открыть терминал (Cmd, PowerShell, bash).
2. Перейти в папку с скриптом:
   ```bash
   cd /путь/к/папке
   ```
3. Запустить скрипт:
   ```bash
   python script.py  # Или python3 script.py
   ```
**Дополнительно:**
- Если Python не в PATH, указать полный путь к интерпретатору:
  ```bash
  /usr/bin/python3 script.py  # Linux/macOS
  C:\Python39\python.exe script.py  # Windows
  ```
