# **Первое знакомство с Python**

---

### **1. Кто считается первым разработчиком и когда были выпущены версии Python?**
**Ответ:** Python создал **Гвидо ван Россум** в 1991 году. Основные версии:
- Python 1.0 — январь 1994
- Python 2.0 — октябрь 2000
- Python 3.0 — декабрь 2008 (обратно несовместим с Python 2)

---

### **2. Что такое Python? Каковы его основные особенности?**
**Ответ:** Python — высокоуровневый интерпретируемый язык программирования общего назначения.
**Особенности:**
- Простой и читаемый синтаксис
- Динамическая типизация
- Автоматическое управление памятью (сборка мусора)
- Кроссплатформенность
- Большое количество библиотек

---

### **3. Чем отличаются Python 1, Python 2 и Python 3? Почему сейчас используют Python 3?**
**Ответ:**
- **Python 1 (1994)** — базовая версия, минимум возможностей.
- **Python 2 (2000)** — улучшения в Unicode, списковых генераторах.
- **Python 3 (2008)** — несовместим с Python 2, улучшенный Unicode (`str` vs `bytes`).
**Почему Python 3?** Python 2 официально не поддерживается с 2020 года.

---

### **4. Какие популярные IDE и редакторы кода подходят для Python?**
**Ответ:**
- **IDE:** PyCharm, VS Code.

---

### **5. Что такое PATH и зачем добавлять Python в переменные среды?**
**Ответ:**
- **PATH** — системная переменная, содержащая пути к исполняемым файлам.
- **Зачем добавлять Python?** Чтобы запускать Python и `pip` из любой папки в командной строке/терминале.

---

### **6. Что такое pip и для чего он нужен?**
**Ответ:**
- **pip** — пакетный менеджер Python для установки библиотек.
- **Пример использования:**
  ```bash
  pip install numpy  # Установка библиотеки
  pip list          # Список установленных пакетов
  ```
